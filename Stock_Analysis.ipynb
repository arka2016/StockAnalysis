#### Working with Nifty50
Objective:
1 - Get list of Stocks in Nifty 50
2 - Analyze correlation of Closing price with index (last 200 Days)

```python
# Importing python Libraries 
import numpy as np
import pandas as pd
import io
import requests
```

```python
# Get list of stocks in nifty
url = 'https://archives.nseindia.com/content/indices/ind_nifty50list.csv'
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
df
```

```python
tickers=list(df.Symbol)
for i in range(0,len(df.index)):
    tickers[i]=tickers[i]+'.NS'
tickers.append('^NSEI')
```

```python
tickers
```

```python
import yfinance as yf
import datetime as dt
```

```python
start_date=dt.datetime.today()-dt.timedelta(200)
end_date=dt.datetime.today()
```

```python
# This is the angel dataframe 
niftyStocks = yf.download(tickers,start_date,end_date)["Adj Close"]
```

```python
niftyStocks
```

```python
# Plotting the correlation in heatmap
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(niftyStocks.corr(),annot=True)
```
